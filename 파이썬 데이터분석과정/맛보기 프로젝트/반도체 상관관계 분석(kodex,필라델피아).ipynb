```python
df = pd.DataFrame(columns=['date', 'price', 'delta'])

kodex_date=[]
kodex_price=[]
kodex_delta=[]

for k in range(1,362):
    
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(3) # 난수가 생성 생성된 난수에 따라서 요청하는 시간 딜레이
    time.sleep(a)
    
    url = f'https://finance.daum.net/api/quote/A091160/days?symbolCode=A091160&page={k}&perPage=10&pagination=true'

    header = {
        'referer' :'https://finance.daum.net/quotes/A091160?period=day', 
        'user-agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36'
    }

    response = requests.get(url, headers = header)

    data = json.loads(response.text)
    
    print(k," 번째실행중")

    for i in range(10):

        kodex_date.append(data['data'][i]['date'])
        kodex_price.append(data['data'][i]['tradePrice'])


        
        if(data['data'][i]['change'] == 'FALL'):
            kodex_delta.append('-' + str(data['data'][i]['changePrice']))
        elif(data['data'][i]['change'] == 'RISE'):
            kodex_delta.append(data['data'][i]['changePrice'])
        else:
            kodex_delta.append(0)




df['date'] = kodex_date
df['price'] = kodex_price
df['delta'] = kodex_delta

df
```

```python
kodex_delta_list = []
sox_delta_list = []

for k in range(0,3610):
    for i in range(0,4040):
        if(df['date'][k].split(' ')[0] == df2['date'][i].split(' ')[0]):
#             print(df.iloc[k]['date'] , " 코스닥" , df.iloc[k]['delta'], "S&P 500" , df2.iloc[i]['delta'] )
            kodex_delta_list.append(float(df.iloc[k]['delta'])/df.iloc[k]['price'])
            sox_delta_list.append((df2.iloc[i]['delta']/df2.iloc[i]['price']))
    
            break
```
